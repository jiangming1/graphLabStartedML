```python
import graphlab
```

```python
train_comments = graphlab.SFrame('./data/ynu_data/train_first.csv')
```

```python
train_comments.head()
```

```python
train_comments['word_count'] = graphlab.text_analytics.count_words(train_comments['Discuss'])
```

```python
train_comments.head()
```

```python
graphlab.canvas.set_target('ipynb')
```

```python
train_comments['Id'].show()
```

```python
train_comments['Score'].show(view='Categorical')
```

```python
train_comments = train_comments[train_comments['Score'] != 3]
```

```python
train_comments['sentiment'] = train_comments['Score'] >= 4
```

```python
train_comments.head()
```

```python
train_data, test_data = train_comments.random_split(.8, seed=0)
```

```python
sentiment_model = graphlab.logistic_classifier.create(train_data,
                                                      target='sentiment',
                                                      features=['word_count'],
                                                      validation_set=test_data)
```

```python
sentiment_model.evaluate(test_data, metric='roc_curve')
```

```python
sentiment_model.show(view='Evaluation')
```
